## 1. Lists

### Characteristics

- **Ordered**: Elements maintain their order
- **Mutable**: Can add, remove, or modify elements
- **Indexed**: Access elements by position (0-based)
- **Allows duplicates**: Same value can appear multiple times
- **Heterogeneous**: Can contain different types

```python
# Creating Lists

# Empty list
empty_list = []
empty_list2 = list()

# List with elements
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True]
nested = [[1, 2], [3, 4], [5, 6]]

# List comprehension
squares = [x**2 for x in range(10)]

print(f"Numbers: {numbers}")
print(f"Mixed: {mixed}")
print(f"Squares: {squares}")
```

```python
# Common List Operations

my_list = [1, 2, 3, 4, 5]

# Access elements
print(f"First element: {my_list[0]}")
print(f"Last element: {my_list[-1]}")
print(f"Slice [1:4]: {my_list[1:4]}")

# Modify elements
my_list[0] = 10
print(f"After modification: {my_list}")

# Add elements
my_list.append(6)
print(f"After append: {my_list}")

my_list.insert(1, 20)
print(f"After insert: {my_list}")

# Remove elements
my_list.remove(20)
print(f"After remove: {my_list}")

# Search
print(f"Index of 3: {my_list.index(3)}")
print(f"Count of 3: {my_list.count(3)}")
print(f"3 in list: {3 in my_list}")
```

### When to Use Lists

- Need ordered collection
- Need to access elements by index
- Need to modify the collection
- Order matters
- Duplicates are allowed

**Example**: Shopping cart, to-do list, scores in a game

## 2. Tuples

### Characteristics

- **Ordered**: Elements maintain their order
- **Immutable**: Cannot modify after creation
- **Indexed**: Access elements by position
- **Allows duplicates**: Same value can appear multiple times
- **Faster**: Slightly faster than lists for some operations
- **Hashable**: Can be used as dictionary keys (if all elements are hashable)

```python
# Creating Tuples

# Empty tuple
empty_tuple = ()
empty_tuple2 = tuple()

# Single element tuple (comma required!)
single = (1,)

# Tuple with elements
numbers = (1, 2, 3, 4, 5)
mixed = (1, "hello", 3.14, True)

# Without parentheses (tuple packing)
coordinates = 10, 20, 30

# Tuple unpacking
x, y, z = coordinates

print(f"Numbers tuple: {numbers}")
print(f"Coordinates: {coordinates}")
print(f"Unpacked: x={x}, y={y}, z={z}")

# Tuples are immutable
# numbers[0] = 10  # ERROR! Cannot modify tuple
```

```python
# Common Tuple Operations

my_tuple = (1, 2, 3, 4, 5)

# Access elements (same as lists)
print(f"First element: {my_tuple[0]}")
print(f"Last element: {my_tuple[-1]}")
print(f"Slice [1:4]: {my_tuple[1:4]}")

# Search
print(f"Index of 3: {my_tuple.index(3)}")
print(f"Count of 3: {my_tuple.count(3)}")
print(f"3 in tuple: {3 in my_tuple}")

# Concatenation (creates new tuple)
new_tuple = my_tuple + (6, 7)
print(f"Concatenated: {new_tuple}")
```

### When to Use Tuples

- Need immutable ordered collection
- Want to use as dictionary key
- Returning multiple values from function
- Fixed-size collection
- Slightly better performance than lists

**Example**: Coordinates (x, y, z), RGB color values, database records

## 3. Sets

### Characteristics

- **Unordered**: No guaranteed order (though may appear ordered)
- **Mutable**: Can add or remove elements
- **No duplicates**: Automatically removes duplicates
- **Not indexed**: Cannot access by position
- **Fast membership**: O(1) average case for `in` operation
- **Hashable elements only**: Elements must be immutable

```python
# Creating Sets

# Empty set (note: {} creates a dictionary, not a set!)
empty_set = set()  # Correct way
# empty_set = {}  # WRONG! This creates a dictionary

# Set with elements
numbers = {1, 2, 3, 4, 5}
fruits = {"apple", "banana", "orange"}

# From list (removes duplicates automatically)
unique_numbers = set([1, 2, 2, 3, 3, 4])

# Set comprehension
squares = {x**2 for x in range(10)}

print(f"Numbers set: {numbers}")
print(f"Unique from list: {unique_numbers}")
print(f"Squares set: {squares}")
```

```python
# Common Set Operations

my_set = {1, 2, 3, 4, 5}

# Add elements
my_set.add(6)
print(f"After add: {my_set}")

my_set.update([7, 8])
print(f"After update: {my_set}")

# Remove elements
my_set.discard(9)  # No error if not found
my_set.remove(8)   # Raises error if not found
print(f"After remove: {my_set}")

# Membership
print(f"3 in set: {3 in my_set}")
print(f"Length: {len(my_set)}")
```

```python
# Set Operations: Union, Intersection, Difference

set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

union = set1 | set2              # All elements
intersection = set1 & set2       # Common elements
difference = set1 - set2         # In set1 but not set2
symmetric_diff = set1 ^ set2      # In either, not both

print(f"Set1: {set1}")
print(f"Set2: {set2}")
print(f"Union: {union}")
print(f"Intersection: {intersection}")
print(f"Difference (set1 - set2): {difference}")
print(f"Symmetric difference: {symmetric_diff}")
```

### When to Use Sets

- Need unique elements
- Fast membership testing
- Set operations (union, intersection, etc.)
- Order doesn't matter
- Removing duplicates

**Example**: Tracking unique visitors, finding common elements, removing duplicates

## 4. Dictionaries

### Characteristics

- **Key-value pairs**: Store data as key:value pairs
- **Ordered**: Maintains insertion order (Python 3.7+)
- **Mutable**: Can add, remove, or modify key-value pairs
- **Keys are unique**: No duplicate keys (later values overwrite earlier)
- **Fast lookups**: O(1) average case for key access
- **Keys must be hashable**: Keys must be immutable (strings, numbers, tuples)

```python
# Creating Dictionaries

# Empty dictionary
empty_dict = {}
empty_dict2 = dict()

# Dictionary with key-value pairs
student = {
    "name": "Alice",
    "age": 20,
    "grade": "A"
}

# Using dict() constructor
student2 = dict(name="Bob", age=21, grade="B")

# Dictionary comprehension
squares = {x: x**2 for x in range(10)}

print(f"Student: {student}")
print(f"Squares: {squares}")
```

```python
# Common Dictionary Operations

my_dict = {"name": "Alice", "age": 20, "city": "New York"}

# Access values
print(f"Name: {my_dict['name']}")
print(f"Age: {my_dict.get('age')}")
print(f"Salary (with default): {my_dict.get('salary', 0)}")

# Modify values
my_dict["age"] = 21
my_dict["salary"] = 50000  # Add new key-value pair
print(f"After modifications: {my_dict}")

# Remove items
removed = my_dict.pop("salary")
print(f"Removed salary: {removed}")
print(f"After pop: {my_dict}")

# Check membership
print(f"'name' in dict: {'name' in my_dict}")
print(f"'salary' in dict: {'salary' in my_dict}")
```

```python
# Iterating Over Dictionaries

my_dict = {"name": "Alice", "age": 20, "city": "New York"}

# Iterate over keys
print("Keys:")
for key in my_dict:
    print(f"  {key}")

# Iterate over key-value pairs
print("\nKey-Value pairs:")
for key, value in my_dict.items():
    print(f"  {key}: {value}")

# Get all keys, values
print(f"\nAll keys: {list(my_dict.keys())}")
print(f"All values: {list(my_dict.values())}")
```

### When to Use Dictionaries

- Need key-value mapping
- Fast lookups by key
- Storing structured data
- Counting frequencies
- Caching/memoization

**Example**: Student records, word frequency counter, configuration settings

## Detailed Comparison

### Time Complexity Comparison

| Operation | List | Tuple | Set | Dictionary |
|-----------|------|-------|-----|------------|
| Access by index/key | O(1) | O(1) | N/A | O(1) |
| Search (in) | O(n) | O(n) | O(1) | O(1) |
| Insert | O(1)* | N/A | O(1) | O(1) |
| Delete | O(n)* | N/A | O(1) | O(1) |

*Amortized O(1) for append, O(n) for insert/delete at arbitrary position

## Practical Examples

### Example 1: Choosing the Right Structure

```python
# Scenario: Track unique student IDs who attended a class
# Use SET - we need unique values and fast membership checking

attended = {101, 102, 103, 101, 104}  # Duplicate 101 automatically removed
print(f"Attended: {attended}")

# Check if student attended
if 101 in attended:
    print("Student 101 attended")

# Scenario: Store student information (ID -> name mapping)
# Use DICTIONARY - key-value pairs for lookups

students = {
    101: "Alice",
    102: "Bob",
    103: "Charlie"
}

# Fast lookup by ID
name = students[101]
print(f"Student 101: {name}")
```

```python
# Scenario: Maintain ordered list of tasks
# Use LIST - order matters, may have duplicates

tasks = ["homework", "study", "homework", "exercise"]
tasks.append("sleep")
first_task = tasks[0]
print(f"Tasks: {tasks}")
print(f"First task: {first_task}")

# Scenario: Store fixed coordinates
# Use TUPLE - immutable, fixed size

point = (10, 20, 30)  # x, y, z coordinates
x, y, z = point       # Unpack
print(f"Point: {point}")
print(f"Coordinates: x={x}, y={y}, z={z}")
```

### Example 2: Counting Frequencies

```python
# Problem: Count frequency of each word in a sentence

sentence = "the quick brown fox jumps over the lazy dog"

# Approach 1: Using dictionary
word_count = {}
for word in sentence.split():
    word_count[word] = word_count.get(word, 0) + 1

print("Word frequencies:")
for word, count in word_count.items():
    print(f"  {word}: {count}")

# Approach 2: Using Counter (from collections)
from collections import Counter
word_count2 = Counter(sentence.split())
print(f"\nUsing Counter: {word_count2}")
```

### Example 3: Finding Common Elements

```python
# Problem: Find students who took both Math and Science

math_students = {"Alice", "Bob", "Charlie", "David"}
science_students = {"Bob", "David", "Eve", "Frank"}

# Use set intersection
both_classes = math_students & science_students
print(f"Math students: {math_students}")
print(f"Science students: {science_students}")
print(f"Both classes: {both_classes}")
```

## Common Mistakes and Pitfalls

```python
# Mistake 1: Confusing {} for Sets

# WRONG: This creates a dictionary, not a set!
empty = {}  # This is a dict!
print(f"Type of {{}}: {type(empty)}")

# CORRECT: Use set() for empty set
empty_set = set()
print(f"Type of set(): {type(empty_set)}")
```

```python
# Mistake 2: Modifying Tuple

my_tuple = (1, 2, 3)
# my_tuple[0] = 10  # ERROR! Uncomment to see error

# CORRECT: Create new tuple
my_tuple = (10, 2, 3)
print(f"New tuple: {my_tuple}")
```

```python
# Mistake 3: Using List as Dictionary Key

# WRONG: Lists are not hashable
# my_dict = {[1, 2]: "value"}  # ERROR! Uncomment to see error

# CORRECT: Use tuple instead
my_dict = {(1, 2): "value"}
print(f"Dictionary with tuple key: {my_dict}")
```

## Key Takeaways

1. **Lists**: Ordered, mutable, indexed - use when order matters and you need to modify
2. **Tuples**: Ordered, immutable - use for fixed data, dictionary keys, or when immutability is needed
3. **Sets**: Unordered, unique elements - use for fast membership testing and set operations
4. **Dictionaries**: Key-value pairs - use for fast lookups and structured data
5. **Choose wisely**: The right data structure can make your code faster and clearer
6. **Understand trade-offs**: Each structure has different time/space complexities
